```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url = 'https://apfs.dhs.gov/forecast/?component=217&contract_status='
```

```python
# Retrieve page with the requests module
response = requests.get(url)
```

```python
# Create BeautifulSoup object; parse with 'html.parser'
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# Examine the results, use prettify to view nicely
print(soup.prettify())
```

```python
# COLUMN HEADINGS LIST #

precolheads = soup.find('table')
colheads = precolheads.find_all('b')
heads = []
for col in colheads:
    heads.append(col.get_text())

df = pd.DataFrame
df.columns = heads
heads
```
